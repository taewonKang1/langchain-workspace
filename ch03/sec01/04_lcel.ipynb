```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("{language} 할 수 있어?")
```

```python
prompt.invoke("한국말")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model="gemini-3.1-flash-lite-preview",
)
```

```python
chain = prompt | model
```

```python
chain.invoke("한국말")
```

```python
template = "{topic}에 대해 아주 쉽게 300자 내외로 설명해줘"
```

```python
from langchain_core.prompts import PromptTemplate

PromptTemplate.from_template(template)
```

```python
prompt.invoke("Langchain")
```

```python
prompt.stream("Langchain")
```

```python
chain = prompt | model
```

```python
user_input = {"topic": "Langchain"}
```

```python
response = chain.stream(user_input)
```

```python
response
```

```python
for token in response:
    print(token.content[0]["text"], end="", flush=True)
```
