### 메시지 (Messages)

언어 모델(LLM)과의 상호작용에서 문맥(Context)을 구성하는 기본 단위입니다.</br>
모델의 입력 및 출력뿐만 아니라 대화의 상태를 나타내는 데 필요한 내용(Content)과 메타데이터(Metadata)를 모두 전달합니다.</br>
LangChain은 표준 메시지 유형을 제공하여, 호출하는 모델에 관계없이 일관된 동작을 보장합니다.

#### 주요 구성 요소

+ 역할 (Role): 메시지 유형(예: system, user)을 식별합니다.
+ 내용 (Content): 메시지의 실제 내용(텍스트, 이미지, 오디오, 문서 등)을 나타냅니다.
+ 메타데이터 (Metadata): 응답 정보, 메시지 ID, 토큰 사용량 등과 같은 선택적 필드입니다.

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, AIMessage, SystemMessage

model = init_chat_model("google_genai:gemini-2.5-flash")

system_msg = SystemMessage("You are a helpful assistant.")
human_msg = HumanMessage("Hello, how are you?")

# Use with chat models
messages = [system_msg, human_msg]
response = model.invoke(messages)  # Returns AIMessage
```

```python
response
```

#### Message prompts

단순히 텍스트 문자열(string) 하나를 전달하는 것이 아니라 메시지 객체들의 목록(list)을 모델에 전달하는 것입니다.

##### 메시지 프롬프트의 사용처

+ 다중 턴 대화 관리: 여러 차례의 사용자 입력과 AI 응답으로 구성된 대화의 기록(conversation history)을 관리할 때 사용됩니다. 모델은 전달된 전체 메시지 목록을 문맥(context)으로 사용하여 다음 응답을 생성합니다.
+ 시스템 지침 포함: 모델의 행동 방식이나 역할(예: '당신은 시 전문가이다')을 설정하는 SystemMessage를 포함할 때 사용됩니다.
+ 멀티모달 콘텐츠 처리: 텍스트 외에 이미지, 오디오, 파일 등 다양한 형태의 멀티모달 콘텐츠를 포함하여 모델에 전달할 때 사용됩니다.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    # 시스템 메시지: 모델에게 '시(詩) 전문가' 역할을 부여합니다.
    SystemMessage("당신은 시 전문가입니다"),
    
    # 인간 메시지: 사용자 입력으로, '봄에 관한 하이쿠를 작성해달라'는 요청입니다.
    HumanMessage("봄에 관한 하이쿠를 작성해줘"),
    
    # AI 메시지: 모델이 이전에 응답했던 내용으로 대화 기록에 추가됩니다.
    AIMessage("벚꽃이 피어나네...")
]
# 이 메시지 목록을 모델에 전달하여 다음 응답을 얻습니다.
response = model.invoke(messages)
```

```python
response
```

```python
# Dictionary format
messages = [
    {"role": "system", "content": "당신은 시 전문가입니다"},
    {"role": "user", "content": "봄에 관한 하이쿠를 작성해줘"},
    {"role": "assistant", "content": "벚꽃이 피어나네..."}
]
response = model.invoke(messages)
```

#### ToolMessage

LLM이 외부 도구를 호출하여 특정 작업을 수행하도록 지시했을 때, 그 작업이 완료된 후의 결과를 모델에게 전달하여 다음 응답을 생성할 문맥을 완성하는 데 사용됩니다.

```python
from langchain.messages import HumanMessage, AIMessage, ToolMessage

# 1. 모델이 생성한 메시지 (도구 호출을 포함)
# 모델이 직접 답변할 수 없어 'get_weather' 도구를 호출하도록 결정했습니다.
ai_message = AIMessage(
    # 모델의 텍스트 콘텐츠는 비어 있거나, 도구 호출을 설명하는 내용이 들어갈 수 있습니다.
    content=[],
    # 호출할 도구 정보: 이름, 인자, 그리고 고유한 ID를 지정합니다.
    tool_calls=[{
        "name": "날씨_정보_가져오기",
        "args": {"위치": "부산"},
        "id": "call_123"  # 이 ID는 다음 ToolMessage와 일치해야 합니다.
    }]
)

# 2. 도구를 실행하고 결과를 담을 ToolMessage 생성
# 실제 시스템이 '날씨_정보_가져오기' 도구를 실행한 결과입니다.
weather_result = "맑음, 22°C"
tool_message = ToolMessage(
    content=weather_result,         # 도구 실행 결과 (내용)
    tool_call_id="call_123"         # 모델이 요청했던 호출 ID("call_123")와 반드시 일치해야 합니다.
)

# 3. 대화 문맥에 메시지를 추가하여 모델 재호출
messages = [
    HumanMessage("부산의 날씨는 어때요?"), # 사용자 질문
    ai_message,                          # 모델: 날씨 도구 호출 요청
    tool_message,                        # 시스템: 도구 실행 결과 제공
]

# 모델이 이제 도구 실행 결과를 바탕으로 최종 답변을 생성합니다.
response = model.invoke(messages)
```

```python
response.content
```

#### 📦 메시지 내용(Message Content)

`messages` 객체의 구성 요소로, 모델과의 상호작용에서 실제로 전달되는 **데이터 페이로드**를 의미합니다.</br>
이는 단순히 텍스트를 넘어 다양한 형태의 데이터를 포함할 수 있도록 설계되어,</br>
언어 모델의 **멀티모달**(Multimodal) 기능을 지원하는 데 필수적입니다.

##### 1. `content` 속성의 역할

* **실제 데이터:** `content` 속성은 메시지의 실제 내용을 나타냅니다.
* **다양한 형식 지원:** 텍스트 외에 이미지, 오디오, 문서 등 다양한 멀티모달 콘텐츠를 포함할 수 있습니다.

##### 2. 지원되는 내용 형식

`content` 속성은 유연성을 위해 두 가지 주요 형식을 지원합니다.

| 형식 | 설명 | 사용 예시 |
| :--- | :--- | :--- |
| **문자열 (String)** | 가장 기본적인 형태로, 단일 텍스트 프롬프트나 답변을 나타냅니다. | `HumanMessage("안녕하세요.")` |
| **객체 목록 (List of Objects/Dicts)** | 복잡하거나 멀티모달 콘텐츠를 포함할 때 사용됩니다. 각 객체는 **콘텐츠 블록(Content Block)**을 나타냅니다. | `HumanMessage([텍스트 블록, 이미지 블록])` |

##### 3. 멀티모달 콘텐츠와 표준화

LangChain은 다양한 모델 제공자(Provider) 간의 일관성을 위해 콘텐츠 처리를 표준화합니다.

* **멀티모달 (Multimodal):** 텍스트 외에 오디오, 이미지, 비디오, 일반 파일(PDF 등)과 같은 다양한 유형의 데이터와 함께 작업하는 능력을 의미합니다.
* **표준 콘텐츠 블록 (`content_blocks`):** LangChain v1에서는 모델 제공자에 관계없이 작동하는 표준화된 콘텐츠 표현을 위해 `content_blocks`라는 새로운 속성이 메시지에 도입되었습니다. 이 속성은 기존 `content` 속성을 유형 안전(type-safe)하고 표준화된 표현으로 지연 분석(lazily parse)하여 접근하게 합니다.
* **목적:** 개발자는 이 표준 유형을 사용하여 특정 모델 제공자에 종속되지 않고 멀티모달 데이터를 처리할 수 있습니다.


```python
from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, AIMessage, SystemMessage

model = init_chat_model("google_genai:gemini-2.5-flash")

system_msg = SystemMessage("당신은 유능한 비서입니다.")
```

```python
from langchain.messages import HumanMessage

# 1. 문자열 내용 (String Content)
# 가장 일반적인 형태이며, 단일 텍스트 내용을 전달합니다.
human_message_str = HumanMessage("안녕하세요, 잘 지내세요?")
```

```python
human_message_str
```

```python
# Use with chat models
messages = [system_msg, human_message_str]
response = model.invoke(messages)  # Returns AIMessage
```

```python
response
```

```python
# 2. 제공자(Provider) 고유 형식 (List of Dictionaries)
# OpenAI와 같은 특정 모델 제공자의 API 형식에 맞춰 콘텐츠를 딕셔너리 목록으로 전달하는 방식입니다.
human_message_provider_native = HumanMessage(content=[
    # 텍스트 블록
    {"type": "text", "text": "제가 찍은 사진을 봐주세요."},
    # 이미지 URL 블록 (멀티모달 데이터)
    {"type": "image_url", "image_url": {"url": "https://picsum.photos/seed/picsum/200/300"}}
])
```

```python
human_message_provider_native
```

```python
# Use with chat models
messages = [system_msg, human_message_provider_native]
response = model.invoke(messages)  # Returns AIMessage
```

```python
response
```

```python
# 3. 표준 콘텐츠 블록 목록 (LangChain Standard Format)
# LangChain이 모델 제공자에 관계없이 콘텐츠를 표준화하기 위해 사용하는 형식입니다.
# (참고: LangChain v1부터 content_blocks 속성이 추가되었으며, content 인자를 통해 전달될 수 있습니다.)
human_message_standard_blocks = HumanMessage(content_blocks=[
    # 텍스트 블록 (표준 형식)
    {"type": "text", "text": "이 이미지를 분석해 주십시오."},
    # 이미지 블록 (표준 형식)
    {"type": "image", "url": "https://picsum.photos/seed/picsum/200/300"},
])
```

```python
human_message_standard_blocks
```

```python
# Use with chat models
messages = [system_msg, human_message_standard_blocks]
response = model.invoke(messages)  # Returns AIMessage
```

#### 🧩 MessagePlaceholder

ChatPromptTemplate에서 동적인 메시지 목록(List)을 삽입하는 표준적인 방법입니다.

MessagePlaceholder의 핵심 역할은 프롬프트 포맷(formatting) 과정에서 특정 메시지 목록이 렌더링될 위치를 동적으로 지정하는 것입니다.

LangChain의 메시지 객체(BaseMessage 또는 이를 상속받는 HumanMessage, AIMessage 등)의 리스트를 받도록 설계되어 있습니다.

```python
from langchain_core.messages import BaseMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# 1. MessagesPlaceholder 사용을 위한 메시지 리스트 정의 (이전과 동일)
messages = [
    SystemMessage(content="당신은 사용자에게 친절하게 답변하는 언어 모델입니다. 이전 대화를 기억하고 답변하세요."), 
    MessagesPlaceholder(variable_name="history"), # <--- MessagePlaceholder 적용
    HumanMessage(content="{input}") # <--- 현재 질문을 받기 위한 HumanMessage 추가
]

# 2. from_messages를 사용하여 템플릿 생성
prompt_with_placeholder = ChatPromptTemplate.from_messages(messages)

# 3. 체인 구성 (올바른 LCEL 연결)
# 프롬프트(Runnable)와 모델(Runnable)을 연결합니다.
chain = prompt_with_placeholder | model 

# 4. 동적 데이터(history) 준비 (이전과 동일)
sample_history: list[BaseMessage] = [
    HumanMessage(content="내 이름은 김일남이야."),
    AIMessage(content="김일남님 반갑습니다! 어떻게 도와드릴까요?"),
]
# 체인에 전달할 현재 사용자 질문
current_input = "내 이름은 뭐야?"


# 5. 체인 실행 예시 (invoke 시 입력 딕셔너리 전달)
print("--- 체인 실행 시작 (MessagePlaceholder에 history 전달) ---")

# invoke의 입력(딕셔너리)에 MessagePlaceholder와 HumanMessage 변수 전달
result = chain.invoke({
    "history": sample_history,  # MessagesPlaceholder의 'history' 변수에 대화 기록 리스트 전달
    "input": current_input      # HumanMessage의 'input' 변수에 현재 질문 문자열 전달
})

# 응답 출력 (모델 응답은 BaseMessage 객체이므로 content를 추출)
print(f"\n[최종 LLM 응답]:\n{result.content}")
```
