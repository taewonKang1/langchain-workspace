```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain.tools import tool

@tool
def add(a: int, b: int) -> int:
    """
    두 숫자를 더해 그 합을 반환합니다.
    """
    return a + b

@tool
def subtract(a: int, b: int) -> int:
    """
    첫 번째 숫자에서 두 번째 숫자를 빼기합니다.
    """
    return a - b

@tool
def multiply(a: int, b: int) -> int:
    """
     두 숫자를 곱하고 그 결과를 반환합니다.
    """
    return a * b

@tool
def divide(a: int, b: int) -> float | str:
    """
     첫 번째 숫자를 두 번째 숫자로 나눕니다.
     0으로 나누려고 하면 에러 메시지를 반환합니다.
    """
    if b == 0:
        return "Error: 0으로는 나눌 수 없습니다."
    return a / b
```

```python
tools = [add, subtract, multiply, divide]
```

```python
from langchain.agents import create_agent

agent = create_agent(model="google_genai:gemini-3-flash-preview", tools=tools)
```

```python
from langchain.messages import SystemMessage, HumanMessage

agent.invoke({
    "messages": [HumanMessage(content="1 + 1은 뭔가요?")]
})
```

```python
from langchain.messages import SystemMessage, HumanMessage

agent.invoke({
    "messages": [HumanMessage(content="123 + 456 * 789 / 123은 뭔가요?")]
})
```
