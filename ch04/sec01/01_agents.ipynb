```python
from langchain.tools import tool

# [도구 정의 방법]
# 1. 타입 힌팅
# 2. Docsting
# 3. tool 데코레이션
@tool
def get_weather(city: str) -> str:
    """특정 도시의 날씨를 가져옵니다."""
    return f"{city}: 맑음"
```

```python
from langchain.agents import create_agent

agent = create_agent(model="google_genai:gemini-3-flash-preview", tools=[get_weather])
```

```python
from langchain.messages import SystemMessage, HumanMessage

agent.invoke({
    "messages": [HumanMessage(content="내 이름은 김일남이야.")]
})
```

```python
from langchain.messages import SystemMessage, HumanMessage

agent.invoke({
    "messages": [HumanMessage(content="부산의 날씨를 알려주세요.")]
})
```
